## Using Math

To use math directly, we'll calculate number of possible permutations without order from original teams and subtract `len(distinct_matches)` from it. 

The number of ways to choose $k=2$ teams from $N=211$ teams is 

$$ \binom{N}{k} = \frac{N!}{(N-k)!k!} = \frac{211!}{(211-2)!2!}=\frac{211!}{209!2!}$$

And then we subtract all matches that happened from it. Calling it $M$ (which values $5527$), the answer is

$$ \frac{N!}{(N-k)k!} - M = \frac{211!}{209!2!} - 5527$$

https://en.wikipedia.org/wiki/Combination



```python
memo ={}
def factorial(n):
    """
    Fectorial using memoization.
    """
    if n in memo:
        return memo[n]
    elif n == 0:
        return 1
    else:
        x = factorial(n-1) * n
        memo[n] = x
        return x
```

```python
factorial(2) # 2*1
```

```python
factorial(3) # 3*2*1
```

```python
factorial(4) # 4*3*2*1
```

```python
def solution(N, k, games_that_happened):
    return factorial(N) / factorial(N - k) / factorial(k) - games_that_happened
```

```python
%time

print("matches: ", len(distinct_matches_sorted))
print("teams: ", len(all_teams))

solution(
    N=len(all_teams), 
    k=2, 
    games_that_happened=len(distinct_matches_sorted)
)
```

```python
import re

def get_team_normalized(df, team: str):
    """
    Dado um time 'team', procura as partidas desse time no dataframe 'df'
    tanto quando ele jogou como 'home_team' como 'away_team', normaliza os
    nomes das colunas para não ter diferença entre home/away.
    Args:
        df: pandas DataFrame com os dados de todas as partidas
        team: time para pesquisar.
    Returns:
        Um novo dataframe só com partidas do time 'team'.
    """
    
    home_cols = [c for c in df.columns if 'home' in c]
    away_cols = [c for c in df.columns if 'away' in c]
    
    df_home = (
        df
        [(df.home_team == team)]
        [[c for c in df.columns if c not in away_cols]]
        .rename(columns=lambda x: re.sub('home_', '', x))
    )
    
    df_away = (
        df
        [(df.away_team == team)]
        [[c for c in df.columns if c not in home_cols]]
        .rename(columns=lambda x: re.sub('away_', '', x))
    )
    
    df_team_normalized = pd.concat([df_home, df_away], axis=0).reset_index(drop=True)
    
    return df_team_normalized
```

```python
df_brazil = get_team_normalized(df, team='Brazil')

df_brazil
```

A copa é um evento que acontece de 4 em 4 anos, isso inviabiliza a entrada de dados quentes no nosso dataset, uma vez que 4 anos é muito para a vida de um atleta e um time como o todo. Muitas mudanças podem ocorrer nesse tempo. Como não podemos considerar apenas os amistosos realizados no que antecede a copa, pois seria uma quantidade baixa de dados, a primeira opçao seria:
* organizar o dataset com dados de no maximo 2 copas atrás, incluindo todos os amistosos possiveis. Esses jogos das copas anteriores teriam peso neutro.
* como sempre existe amistosos que antecedem a copa, os amistosos do ano que irá ocorrer a copa, recebem peso maior que os amistosos de copas passadas.
    Exemplo: Amistosos de anos passados = 1.0 
             Jogos da fase de grupos = 1.5 
             Jogos das fases mata-mata = 1.7 (pode ser menor e ir subindo conforme avança e fica mais pŕoximo da final. 
             Amistosos do ano atual = 2.0
O ponto é, vitoria na copa passada não é certeza de vitoria na atual, mas muito provavelmente o antigo vencedor pode ficar entre os vitoriosos da vez.

```python
df_modelo = df[['home_team', 'away_team', 'home_team_fifa_rank', 'away_team_fifa_rank', 
                'home_team_total_fifa_points', 'away_team_total_fifa_points', 'home_team_score', 'away_team_score', 'home_team_result']]
df_modelo.describe()
```

```python
from sklearn.model_selection import train_test_split

#Separando as features e as variaveis
X= df_modelo.drop('home_team_result', axis=1)
y= df_modelo['home_team_result']

#dividindo os dados em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=38)
```

```python
#mó vacilo, as variaveis são categoricas (strings), o modelo só
# aceita dados numericos
#isso aqui deve resolver o problema
# X_train = pd.get_dummies(X_train, columns=["home_team", "away_team"])
#nao resolveu hehe
```

```python
# Esse bloco aqui considera a mesma ideia do dummies,
# mas agora utilizamos o LabelEncoder para fazer isso
from sklearn.preprocessing import LabelEncoder

cols = ['home_team', 'away_team']

for col in cols:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col] = le.transform(X_test[col])
```

```python
X_train
```

```python
X_train.describe()
```

```python
from sklearn.svm import SVC
svm=SVC()
#isso instancia ele

#partiu treino
svm.fit(X_train, y_train.ravel())
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


#iniciando previsões com o model treinado
y_pred = svm.predict(X_test)

#calculo da accuracy
acc = accuracy_score(y_test, y_pred)
print("Acuracia: ", acc)

#calculo da precisao do modelo
precisao = precision_score(y_test, y_pred, average='weighted')
print("Precisão: ",precisao)
"""Isso vai retornar a média do score da validação cruzada. É util para
lidar com conjuntos de dados desbalanceados"""

#calculo da porcentagem de verdadeiros positivos (revocação)
rec = recall_score(y_test, y_pred, average='weighted')
print("Revocação: ", rec )
"""Em outras palavras, é a capacidade do modelo de identificar corretamente
as instancias positivas da classe. É importante em problemas onde o
objetivo é identificar positivos com alta precisao.
Um 'positivo' é uma classe especifica que o modelo marca como verdadeira
nesse caso, poderia ser as ocorrencias que são identificadas corretamente
pelo modelo"""

#calculando f1 média harmonica entre precisao e recall
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1: ",f1 )

#plot matriz de confusão
mat = confusion_matrix(y_test, y_pred)
print("Matriz de confusão:\n", mat)

"""A matriz de confusão é uma tabela que mostra as previsões corretas e
erradas do modelo. Ela é composta por 4 valores: 
Verdadeiros positivos (VP)
Falsos positivos (FP)
Verdadeiros negativos (VN)
Falsos negativos (FN)

VP representa o numero de previsoes corretas de classe positiva
VN o numero de previsoes corretas da classe negativa
FN o numero de previsoes erradas da classe negativa
Esses valores são úteis para entender a performance do modelo e 
identificar possíveis problemas, como desequilíbrios de classes.
"""
```

<h1> Random Forest

Durante esse primeiro uso do Random Forest, utilizaremos as mesmas features que foram utilizadas no SVM, afim de apenas visualizar se na primeira instancia do modelo, a escolha de qual modelo utilizar já consegue trazer diferenciações na performance.

```python
modelo_rf = df[['home_team', 'away_team', 'home_team_fifa_rank', 'away_team_fifa_rank', 
                'home_team_total_fifa_points', 'away_team_total_fifa_points', 'home_team_score', 'away_team_score', 'home_team_result']]
modelo_rf.describe()
```

```python
from sklearn.model_selection import train_test_split

#Separando as features e as variaveis
X = modelo_rf.drop('home_team_result', axis=1)
y = modelo_rf['home_team_result']

#Realizando a divisão de dados em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=48)
```

```python
from sklearn.preprocessing import LabelEncoder

cols = ['home_team', 'away_team']

for col in cols:
    le= LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col] = le.transform(X_test[col])
X_test.head(1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

rf = RandomForestClassifier(n_estimators=500, max_depth=2, random_state=48, oob_score=True)

rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

#iniciando as previsões com Random Forest agora
y_pred = rf.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("Acuracia: ", acc)

precisao = precision_score(y_test, y_pred, average='weighted')
print("Precisão: ", precisao)

rec = recall_score(y_test, y_pred, average='weighted')
print("Revocação: ",rec)

f1 = f1_score(y_test, y_pred, average='weighted')
print("f1: ",f1)

mat = confusion_matrix(y_test, y_pred)
print("matriz de confusão: \n", mat)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.imshow(mat, cmap='Blues')
plt.title("Matriz de confusão")
plt.colorbar()
resultado = ['Vitoria mandante', 'Empate','Derrota mandante']
marks = np.arange(len(resultado))
plt.xticks(marks,resultado, rotation=45)
plt.yticks(marks, resultado)
plt.ylabel("Rotulo real")
plt.xlabel("Rotulo previsto")

for i in range(len(resultado)):
    for j in range(len(resultado)):
        plt.text(j, i, format(mat[i,j], 'd'),
                horizontalalignment='center', color='white' if mat[i,j] > mat.max()/ 2 else "black")
```
